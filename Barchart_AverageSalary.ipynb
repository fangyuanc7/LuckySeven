```python
import pandas as pd
import numpy as np
import math
import plotly.graph_objs as go 
from plotly.offline import init_notebook_mode,iplot
import matplotlib.pyplot as plt
import plotly.plotly as py
```

```python
path = '/Users/shmylm/Desktop/Project/Process/'
```

```python
BA_job = pd.read_csv(path + 'BA_job_clean.csv')
```

```python
BA_job
```

```python
#creat two lists named Lower_Salary and Upper_Salary, which corresponding to the Salary_Lower_Bound 
#and Salary_Upper_Bound columns in the state dataframe. Append value 0 to the list if the data is missing
Lower_Salary = []
Upper_Salary = []

for i in range(len(BA_job)):
    if math.isnan(BA_job['Salary_Lower_Bound'][i]) == True:
        Lower_Salary.append(0)
    else:
        Lower_Salary.append(int(BA_job['Salary_Lower_Bound'][i]))
        
for i in range(len(BA_job)):
    if math.isnan(BA_job['Salary_Upper_Bound'][i]) == True:
        Upper_Salary.append(0)
    else:
        Upper_Salary.append(int(BA_job['Salary_Upper_Bound'][i]))
```

```python
#Get the estimated average salary for each position
Average_Salary = []
for i in range(len(Lower_Salary)):
    Average_Salary.append((Lower_Salary[i] + Upper_Salary[i])/2)
```

```python
#Create a new column named Average_Salary in BA_job dataframe
BA_job['Average_Salary'] = Average_Salary
```

```python
Grouped_BA = BA_job.groupby('State')
```

```python
Grouped_BA.describe()
```

```python
# Create data used for barchart
State_name = Grouped_BA.describe().index.get_values()
Average_Mean = Grouped_BA.describe()['Average_Salary']['mean']
```

```python
# Create data and layout objects and get graph
data = [go.Bar(
        x = State_name,
        y = Average_Mean,
        width = 1,
        textposition = 'auto',
        marker = dict(
        color = 'rgb(158,202,225)',
        line = dict(
        color = 'rgb(8,48,107)',
        width = 1.5),
        opacity = 0.6,
       ))]
layout = dict(title = 'Average Salary for Business Analyst Positions in Different States',
             xaxis = dict(
             ticklen = 5,
             zeroline = False,
             gridwidth = 2,
             ),
             yaxis = dict(
             ticklen = 5,
             gridwidth = 2,
             ),)

fig = go.Figure(data = data, layout = layout)
py.iplot(fig, filename = 'basic-bar')
```
