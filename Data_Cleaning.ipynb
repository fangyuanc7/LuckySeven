```python
def get_range(df):
    lower = []
    upper = []
    for i in range(len(df['Salary'])):
        if type(df['Salary'][i]) != str:
            if math.isnan(df['Salary'][i]) == True:
                lower.append(None)
                upper.append(None)
        else:
            salary = df['Salary'][i]
            salary = salary.split('k')
            lower.append(salary[0][1:])
            upper.append(salary[1][2:])
    return lower, upper
```

```python
BA_job['Salary_Lower_Bound'] = get_range(BA_job)[0]
BA_job['Salary_Upper_Bound'] = get_range(BA_job)[1]
```
