```python
def modify_data(df):

    for i in range(len(df['Salary'])):
        if type(df['Salary'][i]) != str:
            if math.isnan(df['Salary'][i])==True:
                df['Salary'][i]=df['Salary'][i]
        elif 'per hour' not in df['Salary'][i]:
            df['Salary'][i]=df['Salary'][i]
        elif 'per hour' in df['Salary'][i]:
            l=int(int(df['Salary'][i][1:3])*40*4*12/1000)
            u=int(int(df['Salary'][i][5:7])*40*4*12/1000)
            df['Salary'][i]='$'+str(l)+'k'+'-'+'$'+str(u)+'k'
    return df
        
        
```

```python
new_BA=modify_data(BA_job)
```

```python
def get_range(df):
    lower = []
    upper = []
    for i in range(len(df['Salary'])):
        if type(df['Salary'][i]) != str:
            if math.isnan(df['Salary'][i])==True:
                lower.append(None)
                upper.append(None)
        else:
            salary = df['Salary'][i]
            salary = salary.split('k')
            lower.append(salary[0]+'k')
            upper.append(salary[1][1:]+'k')
    return lower,upper
```

```python
get_range(new_BA)
```

```python
BA_job['Salary_Lower_Bound']=get_range(new_BA)[0]
BA_job['Salary_Upper_Bound']=get_range(new_BA)[1]
```
