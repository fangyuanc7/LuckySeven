```python
import pandas as pd
import numpy as np
import glob
```

```python
path = r'/Users/shmylm/Desktop/Project/fetched' 
allFiles = glob.glob(path + "/*.csv")

list_ = []

for file_ in allFiles:
    df = pd.read_csv(file_, index_col = None, header = 0)
    list_.append(df)

BA_job = pd.concat(list_, axis = 0, ignore_index = True)
```

```python
BA_job.to_csv('BA_job.csv', index = False)
```
